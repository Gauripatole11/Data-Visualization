```python
import seaborn as sns
import pandas as pd

# Sample data (replace with your actual data)
data = {'survived': [0,1,1,0],
        'pclass': [1,2,3,None],
        'sex': ['male', 'female', 'female', 'male'],
        'age': [20,None,30,22],
        'sibsp': [1,1,0,None],
        'parch': [0,0,None,1],
        'fare': [7.25, None, 8.05, 30.5],
        'embarked': ['S','C',None,'S']}
df = pd.DataFrame(data)

# Count missing values
missing_values = df.isnull().sum()

# Create a bar chart with Seaborn
sns.barplot(x=missing_values.index, y=missing_values)

# Customize the chart
plt.xlabel('Column Name')
plt.ylabel('Number of Missing Values')
plt.title('Missing Values Distribution in Titanic Dataset')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset from seaborn
titanic = sns.load_dataset("titanic")

# Calculate the correlation matrix
correlation = titanic.corr()

# Create the heatmap with coolwarm colormap
heatmap = sns.heatmap(correlation, cmap="coolwarm", annot=False, fmt=".5f", linewidths=0.0)

# Rotate x-axis labels
plt.xticks(rotation= 360)  # Correct indentation for xaxis rotation

# Customize the heatmap with a title 
heatmap.set_title('titanic.corr()')  # Add title

plt.show()
```
